```python
#import required libraries
import nltk
from nltk import sent_tokenize
from nltk import word_tokenize
import string
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.probability import FreqDist
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

### Building Corpus

```python
for i in range(1,21):
    name = "doc"+str(i)
    f = open(name+".txt", "a")
    content = input("Enter the content: ")
    f.write(content)
    f.close()
```

### Preprocessing the documents

```python
stop_word = set(stopwords.words('english'))

def tokenization(doc):
    doc = doc.translate(str.maketrans('', '', string.punctuation))
    words = word_tokenize(doc)
    return words

def stopWord(words):
    aftersw = []
    for l in words:
        if l not in stop_word:
            aftersw.append(l)
            
    return aftersw

def stemming(aftersw):
    ps = PorterStemmer()
    stemwords = []
    
    for a in aftersw:
        stemwords.append(ps.stem(a))
        
    return stemwords

def caseConversion(stemwords):
    lowerWords = []
    for w in stemwords:
        lowerWords.append(w.lower())
    return lowerWords

vocab = []
processed_doc = []
```

```python
for i in range(1,21):
    f = open("doc"+str(i)+".txt", "r")
    doc = f.read()
    
    #print("For doc",i)
    words = tokenization(doc)
    #print("After word tokenization\n",words)

    aftersw = stopWord(words)
    #print("\nAfter stop words removal\n",aftersw)

    stemwords = stemming(aftersw)
    #print("\nAfter stemming\n",stemwords)

    lowerWords = caseConversion(stemwords)
    #print("\nAfter case conversion\n",lowerWords)
    
    vocab.extend(lowerWords)
    
    processed_text = ' '.join(map(str, lowerWords))
    
    print(processed_text)
    processed_doc.append(processed_text)
    f.close()
```

### Tf-Idf

```python
vectorize = TfidfVectorizer(vocab)

vectorize.fit(processed_doc)
response = vectorize.transform(processed_doc)
```

```python
datatfidf = pd.DataFrame(response.toarray(), columns=vectorize.get_feature_names())
datatfidf
```

### Query processing

```python
def search_tfid(k, query):
    
    words = tokenization(query)
    aftersw = stopWord(words)
    stemwords = stemming(aftersw)
    lowerWords = caseConversion(stemwords)
    vocab.extend(lowerWords)
    processed_query = ' '.join(map(str, lowerWords))
    
    #print(processed_query)
    queryVectorize = vectorize.transform([processed_query])
    
    cosine = cosine_similarity(response,queryVectorize)
    cosine = cosine.reshape((-1,))
    
    result = []
    ranking = cosine.argsort()[-k:][::-1]
    print(ranking)
    
    for i in ranking:
        result.append(("doc"+str(i)+".txt",cosine[i]))
        
    return result
```

```python
query = "entertainment and literature"
results = search_tfid(5,query)
results
```
