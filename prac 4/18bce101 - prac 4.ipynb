Document Clustering

Perform document clustering.
 
Input: Document collection (without class information)
Output: Cluster of documents 


```python
#importing necessary libraries

from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import cosine_similarity
```

#### Reading the docs into a list

```python
doc = []
title = []

for i in range(1,21):
    f = open("doc"+str(i)+".txt", "r")
    doc.append(f.read())
    s = "doc" + str(i)
    title.append(s)
    
doc
```

#### Finding the TF-IDF vector form

```python
vectorize = TfidfVectorizer(stop_words='english')
vectorize.fit(doc)
response = vectorize.transform(doc)

datatfidf = pd.DataFrame(response.toarray(), columns=vectorize.get_feature_names())
datatfidf
```

#### Applying kmeans to the document vectors

```python
km = KMeans(n_clusters=3)
km.fit(response)
labels = km.labels_.tolist()

print(labels)
df_cluster = pd.DataFrame(list(zip(title, labels)),columns=['Title', 'Label'])
```

```python
df_cluster.sort_values(by=['Label'])
```

#### Visualization and observations

```python
plt.scatter(df_cluster['Title'], df_cluster['Label'], c=labels, cmap='viridis')
plt.xticks(rotation=90)
plt.show()
```

```python
vals = len(set(labels))
order_centroids = km.cluster_centers_.argsort()[:, ::-1]
terms = vectorize.get_feature_names()

print("Top terms per cluster:")
print("---------------------------------------------------------------------\n")

for i in range(vals):
    print("Cluster %d:" % i),
    for ind in order_centroids[i, :5]:
        print('\t\t%s' % terms[ind])
```

#### Finding cosine similarities

```python
cosine_similarities_docs = cosine_similarity(response)
df_cosine_sims = pd.DataFrame(cosine_similarities_docs,index=title, columns=title)
df_cosine_sims
```
