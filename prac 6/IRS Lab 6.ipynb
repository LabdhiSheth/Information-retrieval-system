```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
import regex
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
from scipy.linalg import svd
from numpy import diag
from numpy import dot
```

```python
# Documents of Courpus
Document = {'D1' : 'I like sports very very much!',
           'D2' : 'Badminton is my favourite game in all the sports',
           'D3' : 'I like indoor games also. My favourite indoor game is chess',
           'D4' : 'All outdoor games have playground',
           'D5' : 'Hockey is national sport of the India',
           'D6' : 'Sports are good for health',
           'D7' : 'Modiji is belong to political party named as Bhajapa.',
           'D8' : 'All countries have different political parties.',
           'D9' : 'Any one can take part in election of India after 21 year.',
           'D10' : 'Good political parties can buid a good nation.',
           'D11' : 'All politics are not currupted.',
           'D12' : 'In election, every one can vote to any party after age of 18.',
           'D13' : 'My friend want to become a politician.',
           'D14' : 'I like Tarak Mehata Ka Ooltah Chashma show.',
           'D15' : 'My favourite TV channel is Sony SUB.',
           'D16' : 'I like comedy, fighting and romantic movies.',
           'D17' : 'My favourite actor is Amithabh Bachchan.',
           'D18' : 'Entertainment teaches lots of thing in our life.',
           'D19' : 'I watch TV for half an hour in a day for entertainment.',
           'D20' : '3 Idiots is my one of the favourite movie'}



tokens = []

for i in Document:
    tokens.append(word_tokenize(Document[i].lower()))
    


print('After tokenization, the words are:')
print()
print(tokens)
```

```python
# List of stop words
stop_words = stopwords.words('english')

# Stop word removal step
new_tokens = []
for i in range(len(tokens)):
    for j in tokens[i]:
        if j not in stop_words and j.isalnum():
            new_tokens.append(j)
print('After stop word removal, the words are:')
print()
print(new_tokens)
```

```python
# Stemming of the words
terms = []
temp_terms = []
stemmer = PorterStemmer()

for i in new_tokens:
    terms.append(stemmer.stem(i))

terms.sort()
terms_list = terms

for i in terms:
    if i not in temp_terms:
        temp_terms.append(i)

print('Ater stemming, the words are:')
print()
print(temp_terms)
```

```python
vocabulary = []
frequency = []
result = []
idf = []
normalize = []

for d in Document:
    temp_document = []
    w = word_tokenize(Document[d].lower())
    for i in w:
        temp_document.append(stemmer.stem(i))
        
    temp = []
    temp_frequency = []
    for f in temp_terms:
        if f in temp_document:
            temp.append(temp_document.count(f))
            temp_frequency.append(1)
        else:
            temp.append(0)
            temp_frequency.append(0)
    vocabulary.append(temp)
    frequency.append(temp_frequency)
# print(vocabulary)

frequency = np.array(frequency)
vocabulary_array = np.array(vocabulary)
#print(vocabulary_array.sum(axis=1))

for i in range(len(frequency[1])):
    idf.append(math.log2((len(Document)/frequency.sum(axis=0)[i])))
    normalize.append(frequency.sum(axis=0)[i])


for i in range(len(vocabulary_array)):
    result.append((vocabulary_array[i]/vocabulary_array.sum(axis=1)[i]) * idf)
#print(idf)
#print(result)




print('##### TF-IDF Matrix Representation #####')
Data_Representation_TF_IDF = np.matrix(result)

df_TF_IDF = pd.DataFrame(Data_Representation_TF_IDF, columns = temp_terms)
print(df_TF_IDF)
```

```python
query = 'movie'

query_tokens = word_tokenize(query.lower())

query_new_tokens=[]
stop_words = stopwords.words('english')
for i in range(len(query_tokens)):
    if query_tokens[i] not in stop_words and query_tokens[i].isalnum():
        query_new_tokens.append(stemmer.stem(query_tokens[i]))
query_new_tokens.sort()
query_new_tokens_list = query_new_tokens
temp_query_tokens = []

for i in query_new_tokens:
    if i not in temp_query_tokens:
        temp_query_tokens.append(i)
        

#print(temp_query_tokens)

query_vector = []
count = 0
for i in df_TF_IDF:
    if i in temp_query_tokens:
        query_vector.append((query_new_tokens.count(i)/normalize[count])*idf[count])
        count = count + 1
    else:
        query_vector.append(0)
        count = count + 1

query_vector = np.array(query_vector)
similarity = []

query_transpose = query_vector.transpose()
print(query_transpose)
print(query_transpose.shape)

```
